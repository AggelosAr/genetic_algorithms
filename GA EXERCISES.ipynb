**Διασταύρωση ενός σημείου - Single-point crossover**
 <p>A point on both parents' chromosomes is picked randomly, and designated a 'crossover point'. Bits to the right of that point are swapped between the two parent chromosomes. This results in two offspring, each carrying some genetic information from both parents.</p>
 
**Επιλογή γονέων ανάλογη της προσαρμογής  - Roulette Wheel Selection**
<p>In the roulette wheel selection, the probability of choosing an individual for breeding of the next generation is proportional to its fitness, the better the fitness is, the higher chance for that individual to be chosen. Choosing individuals can be depicted as spinning a roulette that has as many pockets as there are individuals in the current generation, with sizes depending on their probability.</p>

```python


crossover_Rate = 0.7
stats  = new_evolution(crossover_Rate)

lines = [str(i) for i in range(1, len(stats)+1)]

stats = np.array(stats)

print("Total iterations : ", int(np.sum(stats[:,1])))
print("Total time       :  {:.3f}".format(np.sum(stats[:,2])), "seconds")

print("\nAverage time per iteration : {:.6f}".format(np.sum(stats[:,2])/np.sum(stats[:,1])), "\n")

print("Average iterations for optimal solution in 20 runs : ", np.mean(stats[:,1]))
print("Average time for optimal solution in 20 runs       :  {:.3f}".format(np.mean(stats[:,2])))

plt.title("Iterations per run in 20 distict runs.", fontdict = font1)
plt.xlabel('runs', fontdict = font2)
plt.ylabel('iterations', fontdict = font2)
plt.plot(lines, stats[:,1], '-bD', c="black", mfc='red')
plt.show()

plt.title("Time per run in 20 distict runs.", fontdict = font1)
plt.xlabel('runs', fontdict = font2)
plt.ylabel('time in seconds', fontdict = font2)
plt.plot(lines, stats[:,2], '-bD', c="black", mfc='red')
plt.show()
```

The above plots were made with the assumption that in the case of no crossover , cloning will be made.
We notice that as we increase the crossover probability, the required iterations needed to find the optimal solution is decreasing and so is the time needed since these two relationships are dependent. This phenomena happens because as we increase the crossover rate the good genes of the best chromosomes in the generation are passed to the new generation increasing the diversity of the population and simultaneously increasing the overall fitness of the said generation.Also the genes of some bad chromosomes are passed as well.As we move forward the good effects of the bigger diversity start propagating giving us these results.
